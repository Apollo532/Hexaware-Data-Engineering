```python
from pyspark.sql import SparkSession
from pyspark import SparkContext

sc = SparkContext.getOrCreate()
spark = SparkSession.builder.appName('CA_Train_DataOps').getOrCreate()

df = spark.read.csv("/content/ca_train_data_final.csv", header=True, inferSchema=True)
df.show(5)
```

```python
df.printSchema()
df.show(5)
```

Selecting, Renaming and Filtering

```python
df.select("CUSTOMER_NBR", "RECENCY_DAYS", "net_sales", "margin").show(5)
```

```python
df.withColumnRenamed("CUSTOMER_NBR", "customer_number")
df.show(5)
```

```python
df.filter(df['RECENCY_DAYS'] > 100).show(5)
```

```python
#Dropping unused columns

df = df.drop("FL_DONT_CALL")
df.show(5)
```

```python
#Sorting
df.orderBy(df['RECENCY_DAYS'].desc()).show(5)
```

```python
#Group By
df.groupBy("CUST_SEGMENT").sum("net_sales", "margin").show(5)
```

```python
#Imputation

df = df.fillna(0)
```

```python
#Built-In Functions

from pyspark.sql.functions import col
```

```python
df.withColumn("Margin_Percentage", col("margin") / col("net_sales")* 100).show(5)
```

```python
from pyspark.sql.functions import when

df.withColumn("Sales_Category", when(df['net_sales'] > 100, "High").otherwise("Low")).show(5)
```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import row_number,rank,percent_rank,ntile

WindowSpec = Window.partitionBy("CUST_SEGMENT").orderBy("RECENCY_DAYS")
df.withColumn("Rank", rank().over(WindowSpec)).show(5)
```
