```python
#Get the total number of employees who have changed their department in the last year from the EmployeeDepartmentHistory table.
query = """
SELECT COUNT(*) AS EmployeeChanges
FROM [HumanResources].[EmployeeDepartmentHistory]
WHERE YEAR(StartDate) = YEAR(GETDATE())-1"""

df = pd.read_sql(query,conn)
df
```

```python
#List all employees who have been in the HumanResources department for more than 2 years.
query = """
SELECT BusinessEntityID, DepartmentID, DATEDIFF(year, StartDate, EndDate) AS YrsInDept
FROM [HumanResources].[EmployeeDepartmentHistory]
WHERE DepartmentID = (SELECT DepartmentID FROM [HumanResources].[Department] WHERE Name = 'Human Resources') AND DATEDIFF(year, StartDate, EndDate) > 2 
"""

df = pd.read_sql(query,conn)
df
```

```python
#Find the number of employees hired in each department during the last year.
query = """
SELECT DepartmentID, COUNT(*) AS HiresBwn0910
FROM [HumanResources].[EmployeeDepartmentHistory]
WHERE YEAR(StartDate) BETWEEN 2009 AND 2010
GROUP BY DepartmentID;"""

df = pd.read_sql(query,conn)
df
```

```python
#Get the total pay for each employee over the past 3 years from the EmployeePayHistory table.
query = """
SELECT BusinessEntityID, SUM(Rate*PayFrequency) AS TotalPay
FROM [HumanResources].[EmployeePayHistory]
WHERE RateChangeDate BETWEEN '2009-01-01' AND '2009-12-31'
GROUP BY BusinessEntityID;"""
df = pd.read_sql(query,conn)
df
```

```python
#List the top 5 employees with the highest pay rates who work in the Sales department.
query = """
SELECT TOP 5 E.BusinessEntityID, E.NationalIDNumber, P.Rate
FROM [HumanResources].[Employee] E
JOIN [HumanResources].[EmployeePayHistory] P ON E.BusinessEntityID = P.BusinessEntityID
JOIN [HumanResources].[EmployeeDepartmentHistory] EDH ON E.BusinessEntityID = EDH.BusinessEntityID
WHERE EDH.DepartmentID = (SELECT DepartmentID FROM [HumanResources].[Department] WHERE Name = 'Sales')
ORDER BY P.Rate DESC;"""

df = pd.read_sql(query,conn)
df
```

```python
#Find the average tenure of employees in each department, ordered by the department with the longest average tenure.
query = """
SELECT DepartmentID, AVG(DATEDIFF(year, StartDate, EndDate)) AS AvgTenure
FROM [HumanResources].[EmployeeDepartmentHistory]
GROUP BY DepartmentID
ORDER BY AvgTenure DESC"""

df = pd.read_sql(query,conn)
df.head()
```

```python
#Get the count of employees grouped by gender and department, including those with no department assigned.

query = """
SELECT e.Gender, edh.DepartmentID, COUNT(e.BusinessEntityID) AS EmployeeCount
FROM [HumanResources].[Employee] e
LEFT JOIN [HumanResources].[EmployeeDepartmentHistory] edh ON e.BusinessEntityID = edh.BusinessEntityID
GROUP BY e.Gender, edh.DepartmentID
"""
df = pd.read_sql(query, conn)
df.head()
```

```python
#Find the top 3 departments where employees have the highest average pay over the past 5 years.

query = """
SELECT TOP 5 edh.DepartmentID, AVG(p.Rate) AS AvgRate
FROM [HumanResources].[EmployeeDepartmentHistory] edh
JOIN [HumanResources].[EmployeePayHistory] p ON edh.BusinessEntityID = p.BusinessEntityID
WHERE p.RateChangeDate >= DATEADD(year, -5, GETDATE())
GROUP BY edh.DepartmentID
ORDER BY AvgRate DESC
"""
df = pd.read_sql(query, conn)
```

```python
#Retrieve employees who have worked in more than one department.

query = """
SELECT BusinessEntityID, COUNT(DISTINCT DepartmentID) AS DepartmentCount
FROM [HumanResources].[EmployeeDepartmentHistory]
GROUP BY BusinessEntityID
HAVING COUNT(DISTINCT DepartmentID) > 1
"""
df = pd.read_sql(query, conn)
df.head()
```

```python
#Find the total pay spent per department over the last 2 years.

query = """
SELECT edh.DepartmentID, SUM(p.Rate * p.PayFrequency) AS TotalPay
FROM [HumanResources].[EmployeePayHistory] p
JOIN [HumanResources].[EmployeeDepartmentHistory] edh ON p.BusinessEntityID = edh.BusinessEntityID
WHERE p.RateChangeDate >= DATEADD(year, -2, GETDATE())
GROUP BY edh.DepartmentID
"""
df = pd.read_sql(query, conn)
df.head()
```

```python
#List the employees who have had the most pay rate changes in the last 5 years.


query = """
SELECT TOP 5 BusinessEntityID, COUNT(*) AS PayChanges
FROM [HumanResources].[EmployeePayHistory]
WHERE RateChangeDate >= DATEADD(year, -5, GETDATE())
GROUP BY BusinessEntityID
ORDER BY PayChanges DESC"""

df = pd.read_sql(query,conn)
df.head()
```

```python
#Rank employees by their cumulative pay over their entire employment period in the company.

query = """
SELECT BusinessEntityID, SUM(Rate * PayFrequency) AS TotalPay,
RANK() OVER (ORDER BY SUM(Rate * PayFrequency) DESC) AS PayRank
FROM [HumanResources].[EmployeePayHistory]
GROUP BY BusinessEntityID
"""
df = pd.read_sql(query, conn)
df.head()
```
