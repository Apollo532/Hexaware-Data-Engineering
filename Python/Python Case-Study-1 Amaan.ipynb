```python
#17. Write a recursive function to find all employees reporting under a specific manager in the Employee DataFrame.
# (Manager data missing, hence subordinates calculated by OrganizationLevel, where there is no single level for single person, hence we can
# fetch lower level employees without recursion).


def find_subs(employee_df, employee_id):
    upper_level = employee_df[employee_df['BusinessEntityID'] == employee_id]['OrganizationLevel'].values[0]
    subs = employee_df[employee_df['OrganizationLevel'] < upper_level]
    subs = subs.sort_values('OrganizationLevel', ascending = False)
    return subs

employee_df[employee_df['BusinessEntityID'] == 121]['OrganizationLevel'] #3.0
find_subs(employee_df, 121)
```

```python
#18. Rank employees by HireDate
employee_df['HireDateRank'] = employee_df['HireDate'].rank()

employee_df[['BusinessEntityID', 'HireDate', 'HireDateRank']].sort_values('HireDateRank')
```

```python
#19. Define a function to retrieve employee details based on DepartmentID and HireDate.

def get_details(employee_df, dept_history_df, department_id, hire_date):
    dept_emps = dept_history_df[dept_history_df['DepartmentID'] == department_id]
    merged = pd.merge(employee_df, dept_emps, on='BusinessEntityID', how='inner')
    details_df = merged[merged['HireDate'] == hire_date]
    return details_df


employee_details = get_details(employee_df, dept_history_df, department_id=3, hire_date='2013-05-30')
employee_details
```

```python
#20. Find employees hired in the last 30 days
#Data is old, hence taking sample date from which 30 day period is calculated - current date can be used by pd.Timestamp.now()

thirty_days_ago = pd.to_datetime('2013-06-01') - pd.DateOffset(days=30)
hired = employee_df[employee_df['HireDate'] >= thirty_days_ago]

hired
```

```python
#21. Calculate the retention rate of employees over the past 5 years - no EndDate available in employee, hence using dept_history_df
five_years_ago = pd.to_datetime('2013-12-31') - pd.DateOffset(years = 5)

dept_history_df['StartDate'] = pd.to_datetime(dept_history_df['StartDate'])
dept_history_df['EndDate'] = pd.to_datetime(dept_history_df['EndDate'])
                                            
hired_5_years_ago = dept_history_df[dept_history_df['StartDate'] <= five_years_ago]
retained_employees = hired_5_years_ago[hired_5_years_ago['EndDate'].isnull()]

retention_rate = len(retained_employees) / len(hired_5_years_ago) * 100

print('Retention Rate is: ',retention_rate)
```
